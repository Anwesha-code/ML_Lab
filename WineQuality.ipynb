Experiment 1: Write a Python program to Load a dataset using Pandas Identify missing values
using isnull() and notnull(). Display the total count of missing values in each column

```python
import pandas as pd

data_red = pd.read_csv("winequality-red.csv", sep=";")
print("Red Wine\n")
print(data_red.head(), "\n")
data_white = pd.read_csv("winequality-white.csv", sep=";")
print("White Wine\n")
print(data_white.head(), "\n")

print("Red shape:", data_red.shape)  
print("White shape:", data_white.shape) 
```

```python
def checkingmissing(df, name):
    print(f"\n{name} Dataset:")
    print("Shape:", df.shape)
    missing = df.isnull()
    print("\nisnull 3 rows:\n", missing.head(3))
    non_missing = df.notnull()
    print("\nnotnull 3 rows:\n", non_missing.head(3))
    missing_count = df.isnull().sum()
    print("\nMissing values count per column:\n", missing_count)
    total_missing = missing_count.sum()
    print(f"\nTotal missing values: {total_missing}")



checkingmissing(data_red, "Red Wine")
checkingmissing(data_white, "White Wine")
```

 Write a program to Read a CSV dataset Remove rows containing missing
values. Compare the dataset size before and after cleaning.

```python
def clean_dataset(df, name):
    print(f"\n {name}")
        
    cleaned = df.dropna()
    print(f"After:  {cleaned.shape[0]} rows \n")
    print(f"Rows dropped: {df.shape[0] - cleaned.shape[0]} \n")
    
    return cleaned

red_cleaned = clean_dataset(data_red, "Red Wine")
white_cleaned = clean_dataset(data_white, "White Wine")

print(f"New shapes - Red: {red_cleaned.shape}, White: {white_cleaned.shape}")
```

Write a Python program to load a dataset with missing values Replace missing
numerical values using fillna() with mean or median. Replace missing categorical values using
appropriate constants.

```python
print("Red Datatypes \n")
print(red_cleaned.dtypes, "\n")
print("white Datatypes \n")
print(white_cleaned.dtypes, "\n")
numerical_cols_red = red_cleaned.select_dtypes(include=['int64', 'float64']).columns
print("Numerical columns IN RED:", numerical_cols_red.tolist(), "\n")
numerical_cols_white = white_cleaned.select_dtypes(include=['int64', 'float64']).columns
print("Numerical columns IN WHITE:", numerical_cols_white.tolist(), "\n")
```

```python
# Calculate mean and median for all numerical columns in both datasets
print("Red wine - Statistics:")
red_means = red_cleaned[numerical_cols_red].mean()
red_medians = red_cleaned[numerical_cols_red].median()
print("Means:\n", red_means)
print("Medians:\n", red_medians)

print("\nWhite wine - Statistics:")
white_means = white_cleaned[numerical_cols_white].mean()
white_medians = white_cleaned[numerical_cols_white].median()
print("Means:\n", white_means)
print("Medians:\n", white_medians)

# Replace missing numerical values with column medians
red_cleaned[numerical_cols_red] = red_cleaned[numerical_cols_red].fillna(red_medians)
white_cleaned[numerical_cols_white] = white_cleaned[numerical_cols_white].fillna(white_medians)

# For categorical columns , fill with mode or 'unknown'
cat_cols_red = red_cleaned.select_dtypes(include=['object']).columns
cat_cols_white = white_cleaned.select_dtypes(include=['object']).columns

if len(cat_cols_red) > 0:
    red_modes = red_cleaned[cat_cols_red].mode().iloc[0]
    red_cleaned[cat_cols_red] = red_cleaned[cat_cols_red].fillna(red_modes)

if len(cat_cols_white) > 0:
    white_modes = white_cleaned[cat_cols_white].mode().iloc[0]
    white_cleaned[cat_cols_white] = white_cleaned[cat_cols_white].fillna(white_modes)

print("\nMissing values handled with medians for numerical and modes for categorical columns.")
```

```python
# Convert quality to integer 
red_cleaned['quality'] = red_cleaned['quality'].astype(int)
white_cleaned['quality'] = white_cleaned['quality'].astype(int)

# Convert other numerical columns to float32 for efficiency
float_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 
              'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 
              'density', 'pH', 'sulphates', 'alcohol']

for col in float_cols:
    red_cleaned[col] = red_cleaned[col].astype('float32')
    white_cleaned[col] = white_cleaned[col].astype('float32')

# Verify data types and check for any remaining missing values
print("Red wine data types after conversion:")
print(red_cleaned.dtypes)
print("\nRed wine missing values:", red_cleaned.isnull().sum().sum())

print("\nWhite wine data types after conversion:")
print(white_cleaned.dtypes)
print("White wine missing values:", white_cleaned.isnull().sum().sum())
```

```python
# Rename columns for better readability
red_cleaned = red_cleaned.rename(columns={
    'fixed acidity': 'fixed_acidity',
    'volatile acidity': 'volatile_acidity', 
    'citric acid': 'citric_acid',
    'residual sugar': 'residual_sugar',
    'free sulfur dioxide': 'free_sulfur_dioxide',
    'total sulfur dioxide': 'total_sulfur_dioxide'
})

white_cleaned = white_cleaned.rename(columns={
    'fixed acidity': 'fixed_acidity',
    'volatile acidity': 'volatile_acidity',
    'citric acid': 'citric_acid', 
    'residual sugar': 'residual_sugar',
    'free sulfur dioxide': 'free_sulfur_dioxide',
    'total sulfur dioxide': 'total_sulfur_dioxide'
})

print("Red wine - First few rows with new column names:")
print(red_cleaned.head())
print("\nWhite wine - First few rows with new column names:")
print(white_cleaned.head())

print("\nRed wine columns:", red_cleaned.columns.tolist())
print("White wine columns:", white_cleaned.columns.tolist())
```

```python
def clean_impossible_values(df, col_medians):
    conditions = {}
    
    conditions['fixed_acidity'] = df['fixed_acidity'] <= 0
    conditions['pH'] = (df['pH'] < 2.8) | (df['pH'] > 4.0)
    conditions['alcohol'] = (df['alcohol'] < 8) | (df['alcohol'] > 14)
    conditions['quality'] = (df['quality'] < 3) | (df['quality'] > 9)
    
    fixed_count = 0
    for col, mask in conditions.items():
        if mask.sum() > 0:
            fixed_count += mask.sum()
            df.loc[mask, col] = col_medians[col]
    
    return fixed_count

red_medians = red_cleaned.median()
white_medians = white_cleaned.median()

print("Issues found before cleaning:")
print("Red - pH out of range:", ((red_cleaned['pH'] < 2.8) | (red_cleaned['pH'] > 4.0)).sum())
print("Red - alcohol out of range:", ((red_cleaned['alcohol'] < 8) | (red_cleaned['alcohol'] > 14)).sum())
print("White - pH out of range:", ((white_cleaned['pH'] < 2.8) | (white_cleaned['pH'] > 4.0)).sum())
print("White - alcohol out of range:", ((white_cleaned['alcohol'] < 8) | (white_cleaned['alcohol'] > 14)).sum())

red_fixed = clean_impossible_values(red_cleaned, red_medians)
white_fixed = clean_impossible_values(white_cleaned, white_medians)

print(f"\nFixed {red_fixed} values in red wine")
print(f"Fixed {white_fixed} values in white wine")

print("\nAll values now in realistic ranges!")
print("Red wine verification passed:", all([
    (red_cleaned['fixed_acidity'] > 0).all(),
    ((red_cleaned['pH'] >= 2.8) & (red_cleaned['pH'] <= 4.0)).all(),
    ((red_cleaned['alcohol'] >= 8) & (red_cleaned['alcohol'] <= 14)).all(),
    ((red_cleaned['quality'] >= 3) & (red_cleaned['quality'] <= 9)).all()
]))

print("White wine verification passed:", all([
    (white_cleaned['fixed_acidity'] > 0).all(),
    ((white_cleaned['pH'] >= 2.8) & (white_cleaned['pH'] <= 4.0)).all(),
    ((white_cleaned['alcohol'] >= 8) & (white_cleaned['alcohol'] <= 14)).all(),
    ((white_cleaned['quality'] >= 3) & (white_cleaned['quality'] <= 9)).all()
]))
```
